## 1. Establish the connection - conecting w/the DB

```python
### Connect to the DB - Establish the connection
import pyodbc

# Valid values for the connection string
driver = '{ODBC Driver 17 for SQL Server}'
server = '(local)'
dbname = 'AdventureWorks2019'
#dbname = 'BikeStores'
user = 'user1'
passwd = 'pass1'

# Construct the Connection String
connection_string = f'DRIVER={driver};SERVER={server};\
    DATABASE={dbname};UID={user};PWD={passwd}'
print('Connection String:\n', connection_string)

# Establish the connection
try:
    connection = pyodbc.connect(connection_string)
    cur = connection.cursor()
except pyodbc.Error as e:
    print('ERROR:', e)
else:
    print('SUCCESS: Connection Established')
```

## 2. Using SQL Server and Quering directly to de DB using cursor

```python
# mk functions to convert SQL queries to DF
import pandas as pd

def df_from_query(qry):     # convert cursor.execute(query) to DF
    cur.execute(qry)
    field_names = [i[0] for i in cur.description]
    get_data = [list(x) for x in cur]
    df = pd.DataFrame(data=get_data, columns=field_names)
    return df
```

```python
### First query, first look to the products table
query1 = ''' SELECT * FROM Production.Product; '''

prods_df = df_from_query(query1)

print(query1)
display(prods_df.iloc[[0, 5, -5, -1]])
prods_df.columns   # to see all cols name of the table cause they are many
# Here I can: 1. see whole data in the Production.Product table; and
# 2. get the table in DF: products_table_df
```

## 3.1. Aggregate function COUNT()
How many products  (COUNT) of each color (GROUP BY) we have? - sorted (ORDER BY)

```python
# Native SQL query direct to the table in the DF
q_1 = ''' SELECT Color, COUNT(Color) as Number
        FROM Production.Product
        GROUP BY Color
        ORDER BY Number DESC'''
df1 = df_from_query(q_1)
display(df1)

# Using Pandas value_counts() - get a sorted series as result
val_count_serie = prods_df.Color.value_counts()
print('-' * 40, '\n', val_count_serie, sep='')

# Using Pandas group by + count() - get unsorted series as result
group_by_serie = prods_df.groupby('Color').Color.count()
print('-' * 40, '\n', group_by_serie, sep='')

# %timeit shows ops to DF are faster but took a time early to read full tbl
```

## 4. Aggregate function w/condition (GROUP BY... HAVING)

```python
## What chip product (<10) have more than 80 orders?
q_3 = ''' SELECT ProductID, COUNT(1) AS count_col
            FROM Purchasing.PurchaseOrderDetail
            WHERE UnitPrice < 10
            GROUP BY ProductID
            HAVING COUNT(1) > 80
            ORDER BY ProductID'''
df3 = df_from_query(q_3)
df3
```

### 5.2. List of products that total invoiced is between x and y $, order by ProdutID or Total Invocied, Asc or Desc:

```python
## Enter range of invoice, x and y.

txt = 'total turnover value'
t_low = 'Lower' + ' ' + txt
t_hig = 'Higher' + ' ' + txt

while True: 
    try:
        x = int(input(f'{t_low}? '))
        y = int(input(f'{t_hig}? '))
        assert x > 0 and y > 0 and x <= y 
    except ValueError as e:
        print(f'ERROR: {e} \n'
              f'{t_low} and {t_hig} must be integers')             
    except AssertionError as e:
        print(f'ERROR: {e} \n'
              f'{t_low} must be less or equal to {t_hig} and both grater than 0')
    except Exception as e:
        print(f'ERROR: {e}')
    else:
        print(f'Total income between {x:,} and {y:,}')
        break
```

```python
## Enter sorting options.

dcol_txs = ('Total_turnover', 'ProductID')
dcol = {'t': (dcol_txs[0], dcol_txs[0]), 'p': (dcol_txs[1], dcol_txs[1])}
dord = {'a': ('Ascendent', 'ASC'), 'd': ('Descendent', 'DESC')}

# Function to make input string:
def inp_trin(dic):
    if dic == dcol:
        o1, o2 = 't', 'p'
    elif dic == dord:
        o1, o2 = 'a', 'd'
    return f'[{o1}: {dic[o1][0]}, {o2}: {dic[o2][0]}]'

# Function to check and complete sorting options:
def sort_inputed(dic, inp):
    if inp.lower() in dic.keys():
        return dic[inp.lower()][1]
    else:
        raise AssertionError
    
while True: 
    try:
        sort_col = input(f'Sorting column {inp_trin(dcol)}? ')
        sort_col = sort_inputed(dcol, sort_col)
        sort_order = input(f'Sorting order {inp_trin(dord)}? ')
        sort_order = sort_inputed(dord, sort_order)  
    except AssertionError as e:
        print(f'ERROR: Invalid Input {e} \n'
              f' Valid Columns: {inp_trin(dcol)} \n'
              f' Valid ordering {inp_trin(dord)}')
    except Exception as e:
        print(f'ERROR: {e}')
    else:
        print(f"Sorted by '{sort_col}' column and '{sort_order}'")
        break
```

```python
## To faster the proofs
x = 0
y = 1_000
sort_col = dcol_txs[0]
sort_order = 'DESC'
```

```python
# The native SQL Query:

q_52 = f''' SELECT ProductID, SUM(LineTotal) AS {dcol_txs[0]}
            FROM Purchasing.PurchaseOrderDetail
            GROUP BY ProductID
            HAVING SUM(LineTotal) >= {x} and SUM(LineTotal) <= {y}
            ORDER BY {sort_col} {sort_order} '''
df_52 = df_from_query(q_52)
df_52
# TO-DO add column w/UnitPrice of the prduct!!!
```

```python
## Native PANDAS - orders_df is the table
# mk the aggregate table w/the total turnover by product
df_52p = orders_df.groupby('ProductID').agg(
    Total_turnover=pd.NamedAgg(column='LineTotal', aggfunc='sum'))
#display(df_52p)
# mk the index as a col to filter and order 
df_52p.reset_index(inplace=True)
#display(df_52p)

# Convert sort options to Pandas needs:
if sort_order == 'ASC':
    sval = True
elif sort_order == 'DESC':
    sval = False

# mk the filter segÃºn sorting options
df_52p1 = df_52p.loc[(df_52p.Total_turnover >= x) & (df_52p.Total_turnover <= y)]\
    .sort_values(by=[sort_col], ascending=sval).reset_index()

del df_52p1['index']

df_52p1

# # The above is enough but to make idem to SQL native:
# df_51p3 = df_51p2.reset_index()
# display(df_51p3)
```
